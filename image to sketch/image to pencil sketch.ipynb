```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('C:/Users/ASUS/Desktop/flower.jpg')
```

```python
cv2.imshow('original image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
```

```python
grey_img=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("new flower", grey_img)
cv2.waitKey(0)
```

```python
inverted_image = 255 - grey_img
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

```python
blurred = cv2.GaussianBlur(inverted_image, (111,111), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(grey_img, inverted_blurred, scale=255.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
